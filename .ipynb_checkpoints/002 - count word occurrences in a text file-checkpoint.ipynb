```python
import os

from collections import defaultdict
from string import punctuation

os.getcwd()
if os.path.exists('helper'):
    file_one = open(os.path.join('helper', 'preamble.txt'), 'r')
    file_two = open(os.path.join('helper', 'moby_dick.txt'), 'r')
```

```python
def word_count(file):
    word_occurrence = defaultdict(lambda:1)

    for line in file:
        for word in line.translate(punctuation).strip().split():
            # remove punctuation so we don't get something like "States," and "States"
            # force lower case so "Me" and "me" aren't considered different
            word_occurrence[''.join(c for c in word if c not in punctuation).lower()] += 1
            
    for key in word_occurrence:
        print(key + ": " + str(word_occurrence[key]))
```

```python
word_count(file_one)
```

```python
word_count(file_two)
```
